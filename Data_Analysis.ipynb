```python
population = population.drop(columns=["Country Code","Indicator Name","Indicator Code"])
population.columns
```

```python
population['Country Name'] = population['Country Name'].str.strip()
population['Country Name'] = population['Country Name'].astype('category')
population['Country Name'].dtype
```

```python
population.isnull().sum()
```

```python
# Transpose the DataFrame so that years become rows and countries become columns
population_transposed = population.set_index('Country Name').T

# Interpolate missing values for each country independently
population_transposed_interpolated = population_transposed.interpolate(method='linear', axis=1)

# Transpose the DataFrame back to the original format
population_interpolated = population_transposed_interpolated.T.reset_index()
population_interpolated
```

```python
X = population["Country Name"].to_numpy().flatten()
y = population.iloc[:,1:]
years = population.columns[1:]
# Replace masked values with NaN
y = np.ma.filled(y, np.nan)
print(y)
print(X)
print(X.shape)
print(y.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming `population` is your DataFrame and 'Country Name' is the first column
countries = population['Country Name'].unique()
fig, axes = plt.subplots(len(countries), ncols=1, figsize=(12, 4 * len(countries)), sharex=True)

# Iterate over each country and create a scatter plot
for i, country in enumerate(countries):
    country_data = population[population['Country Name'] == country].iloc[:, 1:]
    
    # Filter out missing values
    country_data = country_data.replace('--', np.nan).astype(float)
    
    axes[i].scatter(years, country_data.values.flatten())
    axes[i].set_ylabel('Population')
    axes[i].set_title(f'Scatter Plot for {country}')

# Set common xlabel
plt.xlabel('Year')

plt.tight_layout()
plt.show()
```
